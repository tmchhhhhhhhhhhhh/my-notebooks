# Введение в Python

Это первый урок по Python. Мы рассмотрим:
- Переменные
- Типы данных
- Ввод и вывод
- Условия
- Циклы
- Функции


```python
# Переменные и типы данных
x = 10           # целое число (int)
y = 3.14         # число с плавающей точкой (float)
name = 'Alice'   # строка (str)
is_student = True # логический тип (bool)

print(x, y, name, is_student)
```

## Ввод данных

Используем функцию `input` для ввода от пользователя.


```python
age = input('Введите ваш возраст: ')
print('Ваш возраст:', age)
```

## Условия

Используем `if`, `elif`, `else` для ветвления кода.


```python
number = int(input('Введите число: '))
if number > 0:
    print('Положительное')
elif number < 0:
    print('Отрицательное')
else:
    print('Ноль')
```

## Циклы

Пример с циклом `for` и `while`.


```python
# Цикл for
for i in range(5):
    print('For loop iteration:', i)

# Цикл while
count = 0
while count < 5:
    print('While loop iteration:', count)
    count += 1
```

## Функции

Создание функций с помощью `def`.


```python
def greet(name):
    return f'Привет, {name}!'

print(greet('Alice'))
print(greet('Bob'))
```
